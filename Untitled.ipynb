```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
d1 = pd.read_csv('D1.csv')
d1[:6]
```

```python
lr=LinearRegression()
lr.fit(d1[['hedonic_1','hedonic 2','hedonic_3','hedonic 4','functional_1','functional_2','functional_3','functional_4']],d1['affect_1'])
```

```python
lr.coef_
```

```python
error_1=d1['affect_1']-lr.predict(d1[['hedonic_1','hedonic 2','hedonic_3','hedonic 4','functional_1','functional_2','functional_3','functional_4']])
```

```python
plt.hist(error_1)
```
